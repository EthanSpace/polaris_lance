# Import essential packages and functions

```python
import pandas as pd
import numpy as np
import os
```

```python
def table_bins(data_pile, bin_size):
    bins = pd.cut(data_pile, bins = np.arange(1, max(data_pile), bin_size), include_lowest = True)
    counts = bins.value_counts(sort = True)
    table = pd.concat([counts, counts/len(data_pile)], axis = 1, keys = ['counts', 'proportion'])
    return table
```
